# Task 1 : Single Linear Regression

You have learnt that "CPU_frequency" is the parameter with the lowest p-value among the different features of the dataset. Create a single feature Linear Regression model that fits the pair of "CPU_frequency" and "Price" to find the model for prediction.


````python
# Write your code below and press Shift+Enter to execute 
lm = LinearRegression()
lm

X = df[['CPU_frequency']]
Y = df[['Price']]

lm.fit(X,Y)

Yhat = lm.predict(X)
````

Evaluate the Mean Squared Error and R^2 score values for the model. 


````python
# Write your code below and press Shift+Enter to execute 
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

mse = mean_squared_error(df['Price'], Yhat)
print("The mean square error of the price and predicted value is: ", mse)

print(" ")

# r2_score is more logic, it takes Y and Yhat to give the coefficient of determination
r2_score_slr = r2_score(Y, Yhat)
print("The R-square using r2_score for Linear Regression is: ", r2_score_slr)

r2_scorelm = lm.score(X, Y) 
# metrics has the method score that passes X and Y
# probably calculating Yhat in the background to get to r2
print("The R-square using lm.score for Linear Regression is: ", r2_scorelm)
````

# Task 3 - Polynomial Regression
Use the variable "CPU_frequency" to create Polynomial features. Try this for 3 different values of polynomial degrees. Remember that polynomial fits are done using `numpy.polyfit`. 


````python
print("Shape of X:", X.shape)
print("Shape of Y:", Y.shape)

# if X or Y.shape shows an array of 2 numbers, it is not a 1D array
# By flattening Y, you ensure that both X and Y are 1D arrays (using values.flatten)
````

````python
#  Write your code below and press Shift+Enter to execute
# I will try polynomial of 1st, 3rd and 5th order
import numpy as np


X = X.values.flatten() # Based nn shape of X
#np.polyfit fits a polynomial of degree 1 to the data given X (CPU_frequency) and Y (Price)
f1 = np.polyfit(X, Y, 1)

#Creates a polynomial object (p1) from the coefficients f1 returned by np.polyfit(). 
p1 = np.poly1d(f1)

# polynomial degree 3
f2 = np.polyfit(X, Y, 3)
p2 = np.poly1d(f2)

# polynomial degree 5
f3 = np. polyfit(X, Y, 5)
p3 = np.poly1d(f3)
````

Plot the regression output against the actual data points to note how the data fits in each case. To plot the polynomial response over the actual data points, you have the function shown below.


````python
def PlotPolly(model, independent_variable, dependent_variabble, Name):
    x_new = np.linspace(independent_variable.min(),independent_variable.max(),100)
    y_new = model(x_new)

    plt.plot(independent_variable, dependent_variabble, '.', x_new, y_new, '-')
    plt.title(f'Polynomial Fit for Price ~ {Name}')
    ax = plt.gca()
    ax.set_facecolor((0.898, 0.898, 0.898))
    fig = plt.gcf()
    plt.xlabel(Name)
    plt.ylabel('Price of laptops')
````

Also, calculate the R^2 and MSE values for these fits. For polynomial functions, the function sklearn.metrics.r2_score will be used to calculate R^2 values.


````python
from sklearn.metrics import r2_score
````

````python
#  Write your code below and press Shift+Enter to execute

# MSE and R-square of polynomial of order 1
mse_polynomial1 = mean_squared_error(Y, p1(X))
print("The mean square error for the 1st degree polynomial is: ", mse_polynomial1)

r2_score_poly1 = r2_score(Y, p1(X))
print("The R-square value for the 1st degree polynomial is: ", r2_score_poly1)

````

````python
# MSE and R-square of polynomial of order 3
mse_polynomial2 = mean_squared_error(Y, p2(X))
print("The mean square error for the 3rd degree polynomial is: ", mse_polynomial2)

r2_score_poly2 = r2_score(Y, p2(X))
print("The R-square value for the 3rd degree polynomial is: ", r2_score_poly2)

print(" ")
# MSE and R-square of polynomial of order 5
mse_polynomial3 = mean_squared_error(Y, p3(X))
print("The mean square error for the 5th degree polynomial is: ", mse_polynomial2)

r2_score_poly3 = r2_score(Y, p3(X))
print("The R-square value for the 5th degree polynomial is: ", r2_score_poly3)
````

# Task 4 - Pipeline
Create a pipeline that performs parameter scaling, Polynomial Feature generation and Linear regression. Use the set of multiple features as before to create this pipeline. 


````python
#  Write your code below and press Shift+Enter to execute
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import StandardScaler, PolynomialFeatures

#PolynomialFeatures by default is 2
Input = [('Scale', StandardScaler()),('polynomial', PolynomialFeatures(degree = 5, include_bias = False)), ('model', LinearRegression())]
````

````python
pipe = Pipeline(Input)
pipe
````

````python
#First, we convert the data type Z to type float to avoid conversion warnings that may appear as a result of StandardScaler taking float inputs.
#Then, we can normalize the data, perform a transform and fit the model simultaneously.
# Z is the variable containing all the features 
# Y is the price

Z = Z.astype(float)
pipe.fit(Z, Y)

ypipe = pipe.predict(Z)
````

Evaluate the MSE and R^2 values for the this predicted output.


````python
#  Write your code below and press Shift+Enter to execute
mse_pipe = mean_squared_error(Y, ypipe)
print("The MSE for multi-variable polynomial pipeline is: ", mse_pipe)

print("The R^2 for multi-variable polynomial pipeline is: ", r2_score(Y, ypipe))
````
